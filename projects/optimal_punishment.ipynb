
Suppose a crime of societal cost $C_C$ is committed. There are 3 choices of how to deal with the suspect. We can put them through rehab for some time $t$, imprison them for some time $t$, or give them the death penalty. Here is how we will find the cost of each of these decisions.

# Constants
## Costs
\begin{align*}
    C_C &= \text{the societal cost of the crime}
    \\C_R &= \text{the yearly societal cost of rehab for a guilty suspect}
    \\C_{RW} &= \text{the yearly societal cost of rehab for an innocent suspect}
    \\C_P &= \text{the yearly societal cost of imprisoning a guilty suspect}
    \\C_{PW} &= \text{the yearly societal cost of imprisoning an innocent suspect}
    \\C_D &= \text{the societal cost of executing a guilty suspect}
    \\C_{DW} &= \text{the societal cost of executing an innocent suspect}
\end{align*}
## Other constants
\begin{align*}
    r &= \text{the societies discount rate}
    \\p &= \text{the probability the society identifies the correct suspect}
\end{align*}

# Random Variables
\begin{align*}
    L &= \text{remaining lifespan of a criminal}
\end{align*}

# Functions
\begin{align*}
    \lambda_D &= \text{the rate of reoffending given that the punishment is the death penalty}
    \\\lambda_P(t) &= \text{the rate of reoffending given that the punishment is prison for $t$ years}
    \\\lambda_R(a,b) &= \text{the rate of reoffending given that the punishment is rehab for $a$ years after $b$ years of rehab}
\end{align*}

# Options
Expected societal cost of the punishment being rehab for $t$ years and having administered rehab for $b$ years given the criminal has $T$ years left alive:
\begin{equation*}
    U_R(T,t,b) = \begin{cases}
        C_c + p \cdot \left(C_{R}\left(\frac{(1+r)^t-1}{r}\right) + \frac{U_R(T- t-exp(\lambda_R(t,t)),t)}{(1+r)^{t + exp(\lambda_R(t,t))}}\right) + (1-p)\cdot \left(C_{RW}\left(\frac{(1+r)^t-1}{r}\right) + \frac{U_R(T-exp(\lambda_R(t,0)),t,0)}{(1+r)^{exp(\lambda_R(t,0))}}\right) & T > 0
        \\0 & o/w
    \end{cases}
\end{equation*}
Expected societal cost of imprisoning for $t$ years given the criminal has $T$ years left alive:
\begin{equation*}
    U_P(T,t) = \begin{cases}
        C_c + p \cdot \left(C_{P}\left(\frac{(1+r)^t-1}{r}\right) + \frac{U_P(T- t-exp(\lambda_P(t)),t)}{(1+r)^{t + exp(\lambda_P(t))}}\right) + (1-p)\cdot \left(C_{PW}\left(\frac{(1+r)^t-1}{r}\right) + \frac{U_P(T-exp(\lambda_P(t)),t)}{(1+r)^{exp(\lambda_P(t))}}\right) & T > 0
        \\0 & o/w
    \end{cases}
\end{equation*}
Expected societal cost of the death penalty given the criminal has $T$ years left alive:
\begin{equation*}
    U_D(T) = \begin{cases}
        C_c + p \cdot C_D + (1-p)\cdot (C_{DW} + \frac{U_D(T- exp(\lambda_D))}{(1+r)^{exp(\lambda_D)}}) & T > 0
        \\0 & o/w
    \end{cases}
\end{equation*}

# Choice
\begin{equation*}
    \min\left[ \min_t(U_R(L,t,0)), \min_t(U_P(L,t)), U_D(L) \right]
\end{equation*}

# Assumptions
- The first assumption is that innocent people imprisoned/rehabilitated/executed will not commit a crime, thus their costs do not need to be taken into account.
- The effects of rehab reset after each instance of a crime.
- Inter-crime times are exponentially distributed.
- Crimes cannot be commited while in prison or in rehab.
- Criminals will only recommit the same crime they previously committed.

# Analysis
Here is where we will decide, how should we punish the criminal? This is entirely dependent on the parameters, let's go through each:
- $C_c$) As the cost of the crime increases, the cost of a reoccurrence will increase, driving longer sentences (for both rehab and prison). This makes both less desirable to the death penalty. The opposite is also true.
-  As the relative costs of each $C_R, C_P, C_D, C_RW, C_PW, C_DW$  change, the result will change accordingly. As a punishment becomes relatively cheaper compared to it's peers, it will become more favorable. The opposite is also true.
- $r$) The discount rate is an interesting one. As the discount rate increases, the death penalty becomes less favorable, as do long prison/rehab sentences. This is because the future becomes less important in the calculation. This could be seen due to societies expecting growth or just short-sighted members of society. Of course, the opposite is also true, low discount rates indicate the society weighs the future (increasingly) equal to the present. A negative discount rate might even show a society willing to endure costs in the present for their ancestors.
- $p$) As $p$ increases, we start to see closer and closer to what the society truly feels about their options and their criminal's value. A high $p$ will favor the death penalty far more. A low $p$ would likely favor prison sentences over rehab (as the individual is already in prison, there's no reason to rehabilitate them if they are fine).
- $L$) $L$ is an interesting one, it points heavily towards prison sentences in the low end, and points far more towards the death penalty as it increases. More can be said about $L$ in the discussion of $\lambda$.
 - $\lambda$) The relative rates of $\lambda$ are perhaps one of the most important parameters. As $\lambda$ relatively increases for any of the penalties, it becomes far less favorable. Additionally, for rehab or prison, the effect the time parameter has on $\lambda$ makes a huge difference in the behavior of the model. How much experiencing rehab affects your rate of recidivism makes a massive difference to the value of rehab. This also comes into play with $L$, as $\lambda$ becomes relatively larger compared to $L$, the option becomes increasingly more attractive. Interestingly, the death penalty might be the only option that doesn't exhibit that behavior. If all $\lambda$ values increase, the death penalty will seem less useful, as you will save less (relatively) by not having to deal with recidivism.
